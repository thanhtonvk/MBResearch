```python
import matplotlib.pyplot as plt
import cv2
```

```python
x_min,y_min= 530,40
x_max,y_max = 2400,1250

image = cv2.imread('I:/MBBank/Dataset/Signature_UTE/DataRaw/Sig 00001.jpg')
crop = image[y_min:y_max,x_min:x_max]
plt.imshow(crop)
```

```python
import os
ROOT = 'I:/MBBank/Dataset/Signature_UTE/DataRaw/'
list_dir = os.listdir(ROOT)
for file_name in list_dir:
    image = cv2.imread(ROOT+file_name)
    crop = image[y_min:y_max,x_min:x_max]
    cv2.imwrite('./crop/'+file_name,crop)
```
